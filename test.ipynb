```python
# List of experiments

import mlflow

mlflow.set_tracking_uri("https://gviel-mlflow37.hf.space/")
experiments = mlflow.search_experiments()
print("Experiments list:")
for exp in experiments:
    print(f"ID={exp.experiment_id}, name={exp.name}")
```

```python
runs_df = mlflow.search_runs(experiment_ids=[2])  # remplace "0" par ton experiment_id
print(runs_df[['run_id', 'status', 'start_time']])
```

```python
# Dernier run terminé
finished_runs = runs_df[runs_df['status'] == 'FINISHED']
last_run_id = finished_runs.iloc[-1]['run_id']
print("Last finished run ID:", last_run_id)
```
